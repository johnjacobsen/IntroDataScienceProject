```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ANTH = pd.DataFrame(pd.read_csv('./CSVfiles/anthcsv.csv'))
BIOL = pd.DataFrame(pd.read_csv('./CSVfiles/biolcsv.csv'))
CHEM = pd.DataFrame(pd.read_csv('./CSVfiles/chemcsv.csv'))
COGS = pd.DataFrame(pd.read_csv('./CSVfiles/cogscsv.csv'))
CSE = pd.DataFrame(pd.read_csv('./CSVfiles/csecsv.csv'))
ECON = pd.DataFrame(pd.read_csv('./CSVfiles/econcsv.csv'))
HIST = pd.DataFrame(pd.read_csv('./CSVfiles/histcsv.csv'))
LIT = pd.DataFrame(pd.read_csv('./CSVfiles/litcsv.csv'))
MAE = pd.DataFrame(pd.read_csv('./CSVfiles/maecsv.csv'))
MATH = pd.DataFrame(pd.read_csv('./CSVfiles/mathcsv.csv'))
PHYS = pd.DataFrame(pd.read_csv('./CSVfiles/physcsv.csv'))
POLI = pd.DataFrame(pd.read_csv('./CSVfiles/policsv.csv'))
PSYC = pd.DataFrame(pd.read_csv('./CSVfiles/psyccsv.csv'))
VIS = pd.DataFrame(pd.read_csv('./CSVfiles/viscsv.csv'))
#Convert all the csv files into DataFrames
```

```python
Liberalarts= pd.concat([ANTH, ECON, HIST, LIT, POLI, PSYC, VIS])
STEM= pd.concat([BIOL, CHEM, COGS, CSE, MAE, MATH, PHYS])
#Group all the different departments together by category of STEM or Liberal Arts
```

```python
STEM.groupby('Term')['Study Hrs/wk'].mean().plot(y = 'Mean Study Hours per Week', title = 'STEM')
#Look at the raw data through a plot
```

```python
Liberalarts.groupby('Term')['Study Hrs/wk'].mean().plot(y = 'Mean Study Hours per Week', title = 'Liberal Arts')
#Look at the raw data through a plot
#Notice the spike in both STEM and Liberal Arts
#Perhaps the data needs some cleaning
```

```python

coursemeans = [ANTH['Study Hrs/wk'].mean(), ECON['Study Hrs/wk'].mean(), HIST['Study Hrs/wk'].mean(), LIT['Study Hrs/wk'].mean(), POLI['Study Hrs/wk'].mean(), PSYC['Study Hrs/wk'].mean(), VIS['Study Hrs/wk'].mean(), BIOL['Study Hrs/wk'].mean(), CHEM['Study Hrs/wk'].mean(), COGS['Study Hrs/wk'].mean(), CSE['Study Hrs/wk'].mean(), MAE['Study Hrs/wk'].mean(), MATH['Study Hrs/wk'].mean(), PHYS['Study Hrs/wk'].mean()]
#Get a list of the means of all the departments

courses=pd.DataFrame(coursemeans, index = ['ANTH','ECON','HIST','LIT','POLI','PSYC','VIS','BIOL','CHEM','COGS','CSE','MAE','MATH', 'PHYS']).transpose()
#Convert the list into a dataframe and label the columns accordingly
#Since I labelled the rows and have the data in only 1 column I transpose the Frame

courses.plot( title ='Mean Study Hours per Week by Subject', kind='bar')
#Display the mean study hours as a bar plot separated by subject
```

```python
response_rate=0.8
#If I wanted to clean the data further and see the change in results I could easily change this value
STEM = STEM[(STEM['Evals Made']/STEM['Enroll'])>=response_rate]
Liberalarts = Liberalarts[(Liberalarts['Evals Made']/Liberalarts['Enroll'])>=response_rate]
#Remove all Entries with a response rate of 80% or lower
#Hopefully this will remove a lot of the variation caused by low response rates
```

```python
STEM.groupby('Term')['Study Hrs/wk'].mean().plot(y = 'Mean Study Hours per Week', title = 'STEM')
#Visualize the newly cleaned data
```

```python
Liberalarts.groupby('Term')['Study Hrs/wk'].mean().plot(y = 'Mean Study Hours per Week', title = 'Liberal Arts')
#Visualize the newly cleaned data
#We still see the spike but the spike in the data is much smaller now
#It looks like the cleaning worked somewhat
```

```python
ANTH1 = ANTH[(ANTH['Evals Made']/ANTH['Enroll'])>=response_rate]
ECON1 = ECON[(ECON['Evals Made']/ECON['Enroll'])>=response_rate]
HIST1 = HIST[(HIST['Evals Made']/HIST['Enroll'])>=response_rate]
LIT1 = LIT[(LIT['Evals Made']/LIT['Enroll'])>=response_rate]
POLI1 = POLI[(POLI['Evals Made']/POLI['Enroll'])>=response_rate]
PSYC1 = PSYC[(PSYC['Evals Made']/PSYC['Enroll'])>=response_rate]
VIS1 = VIS[(VIS['Evals Made']/VIS['Enroll'])>=response_rate]
BIOL1 = BIOL[(BIOL['Evals Made']/BIOL['Enroll'])>=response_rate]
CHEM1 = CHEM[(CHEM['Evals Made']/CHEM['Enroll'])>=response_rate]
COGS1 = COGS[(COGS['Evals Made']/COGS['Enroll'])>=response_rate]
CSE1 = CSE[(CSE['Evals Made']/CSE['Enroll'])>=response_rate]
MAE1 = MAE[(MAE['Evals Made']/MAE['Enroll'])>=response_rate]
MATH1 = MATH[(MATH['Evals Made']/MATH['Enroll'])>=response_rate]
PHYS1 = PHYS[(PHYS['Evals Made']/PHYS['Enroll'])>=response_rate]
#Individual departments being cleaned

coursemeans1 = [ANTH1['Study Hrs/wk'].mean(), ECON1['Study Hrs/wk'].mean(), HIST1['Study Hrs/wk'].mean(), LIT1['Study Hrs/wk'].mean(), POLI1['Study Hrs/wk'].mean(), PSYC1['Study Hrs/wk'].mean(), VIS1['Study Hrs/wk'].mean(), BIOL1['Study Hrs/wk'].mean(), CHEM1['Study Hrs/wk'].mean(), COGS1['Study Hrs/wk'].mean(), CSE1['Study Hrs/wk'].mean(), MAE1['Study Hrs/wk'].mean(), MATH1['Study Hrs/wk'].mean(), PHYS1['Study Hrs/wk'].mean()]
#Get a list of the means of all the departments

courses1=pd.DataFrame(coursemeans1, index = ['ANTH','ECON','HIST','LIT','POLI','PSYC','VIS','BIOL','CHEM','COGS','CSE','MAE','MATH', 'PHYS']).transpose()
#Convert the list into a dataframe and label the columns accordingly
#Since I labelled the rows and have the data in only 1 column I transpose the Frame

courses1.plot( title ='Mean Study Hours per Week by Subject', kind='bar')
#Display the mean study hours as a bar plot separated by subject
```

```python
liberalmean=Liberalarts['Study Hrs/wk'].mean()
liberalstd=Liberalarts['Study Hrs/wk'].std()
nliberal=Liberalarts['Study Hrs/wk'].count()
#store the mean, sample size, and standard deviation of Liberal Arts study hours per week
```

```python
stemmean=STEM['Study Hrs/wk'].mean()
stemstd=STEM['Study Hrs/wk'].std()
nstem=STEM['Study Hrs/wk'].count()
#store the mean, sample size, and standard deviation of STEM study hours per week
```

```python
significance=scipy.stats.ttest_ind_from_stats(stemmean, stemstd, nstem, liberalmean, liberalstd, nliberal)
#We will calculate the t-value and its corresponding p-value using scipy's ttest function
significance=list(significance)
#convert the results to a list for convenience
pvalue = significance[1]
#The p-value is the second value in the significance list
#Since we are only interested in the p-value for our analysis we will only be printing that
print('Our p-value is ' + str(pvalue))
if pvalue < 0.05:
    print ('Therefore there is a statistically significant relationship between the two variables')
else:
    print('Therefore there is not a statistically significant relationship between the two variables')
#if our pvalue is less than 0.05 then there is a significant relationship
```

```python
avgstudyhrs= pd.DataFrame([stemmean, liberalmean], index = ['STEM', 'Liberal Arts']).transpose()
#Create DataFrame to compare means
avgstudyhrs.plot(title='Mean Weekly Study Hours by major', kind='bar').set_ylim(3,7)
#Create Bar plot to highlight difference in means
```

```python
STEM2= pd.concat([BIOL, COGS, MATH, PHYS, CHEM])
#Now we will do another test without the suspected outliers of MAE and CSE
STEM2 = STEM2[(STEM2['Evals Made']/STEM2['Enroll'])>=response_rate]
#Clean the data like we did in our first test
stem2mean=STEM2['Study Hrs/wk'].mean()
stem2std=STEM2['Study Hrs/wk'].std()
nstem2=STEM2['Study Hrs/wk'].count()
#store the mean, standard deviation, and sample size
```

```python
significance=scipy.stats.ttest_ind_from_stats(stem2mean, stem2std, nstem2, liberalmean, liberalstd, nliberal)
#We will calculate the t-value and its corresponding p-value using scipy's ttest function
significance=list(significance)
#convert the results to a list for convenience
pvalue = significance[1]
#The p-value is the second value in the significance list
#Since we are only interested in the p-value for our analysis we will only be printing that
print('Our p-value is ' + str(pvalue))
if pvalue < 0.05:
    print ('Therefore there is a statistically significant relationship between the two variables')
else:
    print('Therefore there is not a statistically significant relationship between the two variables')
#if our pvalue is less than 0.05 then there is a significant relationship
```
